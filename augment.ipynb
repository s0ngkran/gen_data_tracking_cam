```python
# resized to (720, 720)
for i in range(len(bgs)):
    img = cv2.imread(temp_fol+bgs[i])
    x, y = img.shape[1],img.shape[0]
    if x<y:
        width = 720
        height = int(y/x * 720 )
    else:
        width = int(x/y * 720)
        height = 720

    dim = (width, height)
    resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    img = resized
    img = img[:720,:720]
    cv2.imwrite('augment/bg720_validation/%s.bmp'%str(i).zfill(3),img)
    print(i)
```

# replace

```python
img_fol = 'video_temp/labeled/test/'
for _,__,img_names in os.walk(img_fol):
    print('fin')
bg_fol = 'augment/bg720/'
for _,__,bg_names in os.walk(bg_fol):
    print('fin')
bg_names[:3]
```

```python
for ii in range(3):
    bgi = np.random.permutation(bg_names)
    for i in range(5):
        # load img and bg
        img = cv2.imread(img_fol+img_names[ii])
        bg = cv2.imread(bg_fol+bgi[i])

        # get mask
        c1 = img[:,:,0]==0
        c2 = img[:,:,1]==255
        c3 = img[:,:,2]==0
        mask = c1*c2*c3
        mask.reshape([720,720])
        img[mask] = bg[mask]
        cv2.imwrite('augment/%s_%s.bmp'%(ii,i),img)
    print('tttttttttttttttttttttttttttt')
```
