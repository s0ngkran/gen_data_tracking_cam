```python

frame = copy.copy(frame_ori)
pick = [[100,x] for x in range(200,600,100)]
pick.append([200,600])
pick.append([400,100])
for y, x in pick:
    thres = 30
    picker = frame[y, x]
    print('rgb=', picker, 'thres=', thres)

    red, green, blue = picker
    val_red = frame[:,:,0]
    val_green = frame[:,:,1]
    val_blue = frame[:,:,2]
    c1 = (val_red>red-thres) * (val_red<red+thres)
    c2 = (val_green>green-thres) * (val_green<green+thres)
    c3 = (val_blue>blue-thres) * (val_blue<blue+thres)
    c = c1*c2*c3
    frame[c] = 0
plt.imshow(frame)
plt.show()
```
